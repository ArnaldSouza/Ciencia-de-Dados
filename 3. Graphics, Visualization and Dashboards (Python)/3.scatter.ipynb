```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
base = pd.read_csv('./data/trees.csv')
base.head()
```

```python
# scatter plot considering volume and scatter
plt.scatter(x=base.Girth, y=base.Volume, color = 'green', facecolors = 'none', marker = '*')
plt.title('Trees')
plt.xlabel('Girth')
plt.ylabel('Volume')
```

```python
# line graphic considering volume and girth - without marker
plt.plot(base.Girth,base.Volume)
plt.title('Trees')
plt.xlabel('Girth')
plt.ylabel('Volume')
```

```python
# Scatter plot with jitter
sns.regplot(x=base.Girth,y=base.Volume, data = base, x_jitter = 0.3, fit_reg = False)
```
