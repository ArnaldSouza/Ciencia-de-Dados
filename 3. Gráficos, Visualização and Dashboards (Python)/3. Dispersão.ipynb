```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
base = pd.read_csv('./data/trees.csv')
base.head()
```

```python
# diagram de dispersão considerando volume e a dispersão
plt.scatter(x=base.Girth, y=base.Volume, color = 'green', facecolors = 'none', marker = '*')
plt.title('Trees')
plt.xlabel('Girth')
plt.ylabel('Volume')
```

```python
# gráfico de linha considerando volume e girth
plt.plot(base.Girth,base.Volume)
plt.title('Trees')
plt.xlabel('Girth')
plt.ylabel('Volume')
```

```python
# diagrama de dispersão com jitter
sns.regplot(x=base.Girth,y=base.Volume, data = base, x_jitter = 0.3, fit_reg = False)
```
