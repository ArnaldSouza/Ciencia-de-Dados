Perform an Exploratory Data Analysis process to gain knowledge/understanding of this data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as srn
```

```python
base = pd.read_csv('./data/data.csv', sep=';')
base.shape
```

```python
base.head()
```

```python
srn.histplot(base.PIB, kde=True, bins=10).set(title='PIB')
```

```python
srn.histplot(base.VALOREMPENHO, kde=True, bins=10).set(title='Empenho')
```

```python
grouped = base.sort_values('PIB').head(10)
grouped = grouped.iloc[:,1:3]
grouped
grouped.plot.bar(x='MUNICIPIO',y='PIB', color = 'red')
```

```python
grouped = base.sort_values('VALOREMPENHO').head(10)
grouped = grouped.iloc[:,[1,3]]
grouped
grouped.plot.bar(x='MUNICIPIO',y='VALOREMPENHO', color = 'green')
```
