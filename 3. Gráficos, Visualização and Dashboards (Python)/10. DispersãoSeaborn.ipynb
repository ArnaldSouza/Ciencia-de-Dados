```python
import pandas as pd
import seaborn as srn
import matplotlib.pyplot as plt
```

```python
base = pd.read_csv('./data/co2.csv')
base.head()
```

```python
# Gráfico de dispersão usando os atributos conc e uptake, agrupando por type
srn.scatterplot(data=base, x='conc', y='uptake', hue='Type')
plt.show()
```

```python
# Selecionado registros específicos da base de dados (Quebec/Mississipi)
q = base.loc[base['Type'] == 'Quebec']
m = base.loc[base['Type'] == 'Mississippi']
```

```python
# subgráfico (1 linha e 2 colunas) mostrando gráficos sobre cada região
plt.figure()
plt.subplot(1,2,1)
srn.scatterplot(data=q, x='conc', y='uptake').set_title('Quebec')
plt.subplot(1,2,2)
srn.scatterplot(data=m, x='conc', y='uptake').set_title('Mississippi')
plt.tight_layout()
```

```python
# chilled e non chilled
ch = base.loc[base['Treatment'] == 'chilled']
nc = base.loc[base['Treatment'] == 'nonchilled']
```

```python
# Gráficos com Graphics chilled e non chilled
plt.figure()
plt.subplot(1,2,1)
srn.scatterplot(data=ch,x='conc', y='uptake').set_title('Chilled')
plt.subplot(1,2,2)
srn.scatterplot(data=nc,x='conc', y='uptake').set_title('Non chilled')
plt.tight_layout()
```
