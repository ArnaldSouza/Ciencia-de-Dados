```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
base = pd.read_csv('./data/trees.csv')
base.head()
```

```python
# Gerando boplot
# patch_artist = True -> preenche boxplot com uma cor, showfliers -> outliers
plt.boxplot(base.Volume, vert = False, showfliers = True, notch = False,patch_artist = True)
plt.title('Trees')
plt.xlabel('Volume')
```

```python
# dados por linhae
plt.boxplot(base)
plt.title('Trees')
plt.xlabel('Data')
```

```python
# Criando 3 boxplots, cada um mostrando uma informaçõ diferente
plt.boxplot(base.Volume, vert = False)
plt.boxplot(base.Girth, vert = False)
plt.boxplot(base.Height, vert = False)
plt.title('Trees')
plt.xlabel('Data')
```
