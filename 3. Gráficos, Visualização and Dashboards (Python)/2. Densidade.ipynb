```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
base = pd.read_csv('./data/trees.csv')
base.head()
```

```python
# Histograma com 10 bins e apenas com o prieiro atributo da base de dados
plt.hist(base.iloc[:,1], bins = 10)
```

```python
# Histograma com linha de distibuição de frequqncia, 6 bins
#kde = linha de densidade
sns.histplot(base.iloc[:,1], kde=False, bins=6, color='red').set(title='Trees')
```

```python
#densidade, kdplot método para densidade
sns.kdeplot(base.iloc[:,1], color='blue').set(title='Trees')
```

```python
#densidade e histograma
sns.histplot(base.iloc[:,1], kde=True, bins=10, color='blue').set(title='Trees')
```
